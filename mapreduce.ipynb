```python
def mapper(line):
    return line.strip().split()

def reducer(lines):
    word_count = {}
    for line in lines:
        words = mapper(line)
        for word in words:
            if word in word_count:
                word_count[word] += 1
            else:
                word_count[word] = 1
    return word_count

def map_reduce(input_file):
    with open(input_file, 'r') as f:
        lines = f.readlines()
    
    return reducer(lines)

if __name__ == "__main__":
    word_count = map_reduce('file1.txt')
    for word, count in word_count.items():
        print(f"{word}: {count}")
```

```python
def load_english_words(file_path):
    with open(file_path, 'r') as f:
        return set(word.strip().lower() for word in f)

# Load English words from a file
english_words = load_english_words('english.txt')

def is_english(word):
    return word.lower() in english_words

def mapper(line):
    return line.strip().split()

def reducer(lines):
    non_english_word_count = {}
    for line in lines:
        words = mapper(line)
        for word in words:
            if not is_english(word):
                if word in non_english_word_count:
                    non_english_word_count[word] += 1
                else:
                    non_english_word_count[word] = 1
    return non_english_word_count

def map_reduce(input_file):
    with open(input_file, 'r') as f:
        lines = f.readlines()
    
    return reducer(lines)

if __name__ == "__main__":
    non_english_word_count = map_reduce('file2.txt')
    for word, count in non_english_word_count.items():
        print(f"{word}: {count}")
```
